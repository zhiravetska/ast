```python
import pandas
import pandas as pa
import matplotlib.pyplot as plt 
```

```python
# reading the CSV file
tlf_18 = pandas.read_csv('Total Load Forecast - Year Ahead_2018.csv')
```

```python
df_18 = pa.DataFrame(tlf_18)
```

```python
tlf_19 = pa.read_excel('Total Load Forecast - Year Ahead_2019.xlsx')

df_19 = pa.DataFrame(tlf_19)
```

```python
tlf_20 = pandas.read_csv('Total Load Forecast - Year Ahead_2020.csv')
```

```python
df_20 = pa.DataFrame(tlf_20)
```

```python
tlf_21 = pandas.read_csv('Total Load Forecast - Year Ahead_2021.csv')
```

```python
df_21 = pa.DataFrame(tlf_21)
```

```python
tlf_22 = pandas.read_csv('Total Load Forecast - Year Ahead_2022.csv')
```

```python
df_22 = pa.DataFrame(tlf_22)
```

```python
frames = [df_18, df_19, df_20, df_21, df_22]

result = pa.concat(frames)
df_tot = pa.DataFrame(result)
df_tot
```

```python
df_tot.plot(kind='line', 
        x='Year', 
        y='Forecast max (MW)', 
        color='red') 
  
# set the title 
plt.title('Total year-ahead load forecast in 2018-2022. Latvia') 
  
# show the plot 
plt.show() 
```
